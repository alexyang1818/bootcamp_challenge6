```python
import random

lats = []
lngs = []
for i in range(2000):
    lats.append(random.randint(-90, 89) + random.random())
    lngs.append(random.randint(-180,179) + random.random())

lat_lngs = zip(lats, lngs)
coordinates = list(lat_lngs)
```

```python
from citipy import citipy

cities = []
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1])
    if city not in cities:
        cities.append(city)

# for city in cities:
#     print(f'{city.city_name}, {city.country_code}')
print(len(cities))
```

```python
import requests
import json
from config import openweather_key

openweather_url = 'http://api.openweathermap.org/data/2.5/weather?units=imperial&' # ?q=London,uk&appid={API key}'
weather_list = []

for city in cities:
    city_country = city.city_name + ',' + city.country_code
    url = openweather_url + f'q={city_country}&appid={openweather_key}'
    
    try:
        response = requests.get(url).json()
        # print(json.dumps(response,indent=4))
        city_lat = response['coord']['lat']
        city_lng = response['coord']['lon']
        city_name = response['name']
        country = response['sys']['country']
        max_temp = response['main']['temp_max']
        humidity = response['main']['humidity']
        cloudiness = response['clouds']['all']
        wind_speed = response['wind']['speed']
        description = response['weather'][0]['description']
    except:
        print(f'{city_country} not found. Skipping...')
        continue

    current_weather = {
        'City': city_name,
        'Country': country,
        'Lat': city_lat,
        'Lng': city_lng,
        'Max Temp (F)': max_temp,
        'Humidity (%)':humidity,
        'Cloudiness (%)':cloudiness,
        'Wind Speed (mph)': wind_speed,
        'Current Description': description
    }

    weather_list.append(current_weather)

print(len(weather_list))
```

```python
import pandas as pd

cities_df = pd.DataFrame(weather_list)
cities_df
```

```python
cities_df.to_csv('WeatherPy_Database.csv')
```
